The mapping:
 
- true negative: correctly predicted stock class; 
- false postive: stock class posts that are falsely predicted as wallstreetbet posts;
- false negative: wsb class posts that are falsely predicted as stocks posts.

Base on this mapping, as our project objective is to identify r/stocks posts, the most important metric: Specificity and Negative Predictive Value.
* Specificity (specificity = tn / (tn + fp)), it represents the rate of correctly predicted r/stocks posts among all real r/stocks posts); and 
* Negative Predictive Value (NPV) (npv = tn / (tn + fn)), it represents the rate of true r/stocks posts among all the predicted r/stocks posts). 

Though both Specificity and NPV are important, the NPV is prioritized over Specificity in our casestudy because we want to minimize false negatives as we don't want to falsely target a r/wsb posts for advertisement and ended up channeling more wsb users to our client's app to worsen the current situation (as a trade-off we can afford to miss out some r/stocks posts).

Hence the objective here is to train a model that have the highest NPV score and an reasonably high Specificity score, as well as an reasonably high F1 score.

#### Vectorize the text using TF-IDF 

As TF-IDF transformation has the merit of not only measures the frequency of words but also the relative importance of the words to individual documents, hereby TF-IDF transformation is chose for the first model to set a good foundation for further improvement.

* Define the lemmatizer and customize stop_words list

Although the way TF-IDF vectorizer is set up suggests that there won't be as significant a difference as in CountVectorizer transformer through removing stopwords, here we will still remove the stop words to skip those common words that don't carry meaning for the two subreddtis. 

Later in grid search we will test the model both with and without stop words.

```python
#define lemmatize function 
class LemmaTokenizer:
    def __init__(self):
        self.wnl = WordNetLemmatizer()
    def __call__(self, doc):
        regex_num_ponctuation = '(\d+)|([^\w\s])'
        regex_little_words = r'(\b\w{2,3}\b)'
        return [self.wnl.lemmatize(t) for t in word_tokenize(doc)
                if not re.search(regex_num_ponctuation, t) and not re.search(regex_little_words, t)]
```

```python
# specify for Tf-idf vectorizer to use lemmatized vocabulary for counting  
tvec = TfidfVectorizer(tokenizer=LemmaTokenizer(),stop_words=customized_stop_words)
```

```python
print(tvec.fit_transform(X_train).toarray())
```

```python
X_train_df = pd.DataFrame(tvec.fit_transform(X_train).toarray(), 
                          columns=tvec.get_feature_names_out())
```

```python
# plot top occuring words & compare against CountVectorizer()
X_train_df.sum().sort_values(ascending=False).head(25).plot(kind='barh').invert_yaxis();
```

Using Tf-IDF vectorizer, there are a couple new words showing up among the top 25 unigrams from above results, such as 'yolo','retard', as commpared to results by CountVectorizer as in Notebook Part 03. This is because these words have been important in a small number of posts while absent from the rest of the corpus, hence they are picked up  by the TF-IDF model to be carrying information in differentiating the posts.

#### Model 1: TF-IDF Vectorizer+ Multinormial NB 

* Set up a pipeline with TF-IDF Vectorizer and Multinormial Naive Bayes

```python
pipe_tvec = Pipeline([
    ('tvec', TfidfVectorizer()),
    ('nb', MultinomialNB())
])
```

```python
pipe_tvec_params = {
    'tvec__max_features': [2_000, 3_000, 4_000],
    'tvec__stop_words': [None, 'english',customized_stop_words], # try both with and without stopwords
    'tvec__ngram_range': [(1,1), (1,2),(2,2)] # try unigrams uni+bi-grams, and bigrams
}
```

```python
# Instantiate GridSearchCV.
gs_tvec = GridSearchCV(pipe_tvec,
                        param_grid = pipe_tvec_params, 
                        cv=5, n_jobs=-1) 
```

```python
# Fit GridSearch to training data.
gs_tvec.fit(X_train, y_train)
```

```python
#gs_tvec.best_params_
# commented as it prints out the customized stop_word list which is really long. The other two parameters max_features and ngram_range have been copied below
# {'tvec__max_features': 2000,
# 'tvec__ngram_range': (1, 2),
# 'tvec__stop_words':[....]}
```

```python
%store gs_tvec
```

```python
#F1_score on training dataset
y_pred_train=gs_tvec.predict(X_train)
print('F1_score on the training dataset is:',f1_score(y_pred_train,y_train))

#F1_score on testing dataset
y_pred=gs_tvec.predict(X_test)
print('F1_score on the test dataset is:',f1_score(y_pred,y_test))
```

```python
#accuracy on the training dataset
print('Accuracy on training dataset is:',gs_tvec.score(X_train, y_train))
print('Accuracy on testing dataset is:',gs_tvec.score(X_test, y_test))
```

```python
tn, fp, fn, tp = confusion_matrix(y_test, y_pred).ravel()

spec = tn / (tn + fp)
npv = tn / (tn + fn)

print('Specificity:', spec)
print('Negative Predictive Value:', npv)
```

The NPV score of the Multinomical NB model at around 0.808 is not bad, but the Specificity is not acceptable at about 0.089 (which means the model failed to identify a lot of real r/stocks posts). 

```python
# plot confusion matrix
cm = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=['stock', 'wallstreetbet'])
disp.plot();
```

```python
plot_roc_curve(gs_tvec, X_test, y_test)
plt.plot([0, 1], [0, 1],
         label='baseline', linestyle='--')  
plt.legend();
```

#### Model 5. TF-IDF + Multinomial NB (Oversampling to balanced dataset)

```python
X2_train= df_train_oversample['text']
y2_train = df_train_oversample['label']
```

```python
# check if the training dataset is really balanced 
y2_train.value_counts(normalize=True)
```

* Compared with MultinomialNB in Model 1 without treating the imbalanced dataset, the Specificity has greatly increased from 0.089 to 0.217 with oversampling treatment.
* The perfect Specificity score at 1.00 suggests that the model got 0 false negatives, meaning the model did not wrongly classfifies any r/wallstreetbets post as a target class r/stocks post.
